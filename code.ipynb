```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
source_data_path = r'D:\personal\portfolio\FinDash_manufacture\data.xlsx'
sheet_name = 'cost'
```

```python
cost_data = pd.read_excel(source_data_path, sheet_name=sheet_name)
cost_data.head()
```

```python
selects = ['PLANT', 'PERIOD','TOTAL_COST', 'TOTAL_BUDGET_COST']
group_bys = ['PLANT', 'PERIOD']
display(cost_data.loc[:,selects].groupby(group_bys).sum().style.format('{:,.2f}',))
display(cost_data.loc[:,selects].groupby(group_bys).sum().groupby(['PLANT']).describe().style.format('{:,.2f}',))
```
